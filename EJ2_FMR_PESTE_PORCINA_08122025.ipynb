# **Problema: Gestión de un brote de Peste Porcina Africana (PPA) en España**

Generar propuestas de gestión por diferentes modelos para ser posteriormente evaluada por un juez. Los modelos a emplear serán abierto salvo los frontier de Open AI.

**Contexto:**

España es uno de los mayores productores y exportadores de carne de cerdo del mundo. En el año actual, se confirma un brote de Peste Porcina Africana (PPA) en una explotación de porcino extensivo situada en el suroeste del país, cerca de una zona con alta densidad de jabalíes silvestres.

La enfermedad no afecta a humanos, pero es altamente contagiosa entre suidos y carece de vacuna eficaz. Un manejo inadecuado del brote podría provocar:

    1.Pérdidas económicas masivas en el sector porcino.
    2.Cierre de mercados internacionales.
    3.Impactos sociales en zonas rurales.
    4.Daños a la biodiversidad por medidas de control mal diseñadas.

**Actores implicados:**

    * Ministerio de Agricultura y autoridades autonómicas.
    * Productores porcinos (intensivos y extensivos).
    * Sector exportador.
    * Comunidades rurales.
    * Organizaciones ecologistas.
    * Unión Europea y países importadores.

**Objetivo del problema:**

Diseñar una estrategia integral de respuesta al brote de PPA que permita:

    1. Contener y erradicar la enfermedad.
    2. Minimizar el impacto económico y social.
    3. Cumplir la normativa europea y mantener la credibilidad internacional.
    4. Gestionar el riesgo de transmisión desde fauna silvestre.
    5. Comunicar eficazmente con la población y el sector.

**Tarea para los modelos**

Cada modelo deberá proponer un plan de actuación que incluya, al menos, los siguientes puntos:

    1.Medidas sanitarias inmediatas:
        * Control del foco inicial.
        * Restricciones de movimiento.
        * Gestión de animales infectados y contactos.

    2.Estrategia económica:
        * Compensaciones a ganaderos.
        * Medidas para proteger exportaciones.
        * Apoyo a zonas afectadas.
    
    3.Gestión de fauna silvestre:
        * Control de jabalíes.
        * Prevención de nuevas transmisiones.
        * Consideraciones medioambientales.

    4.Coordinación institucional:
        * Papel de las comunidades autónomas, el Estado y la UE.
        * Cooperación internacional.

    5.Comunicación y percepción pública:
        * Información a la ciudadanía.
        * Mensajes al sector ganadero y a socios comerciales.

    6.Medidas a medio y largo plazo:
        * Prevención de futuros brotes.
        * Cambios estructurales si fueran necesarios.

**Criterios del juez**

El juez evaluará las propuestas según:
    
    1. Eficacia sanitaria (contención y erradicación).
    2. Viabilidad técnica y económica.
    3. Equilibrio entre intereses económicos, sociales y ambientales.
    4. Claridad y coherencia del plan.
    5. Adecuación al contexto español y europeo.
    6. Calidad de la comunicación y gestión del riesgo.


```python
# Cargar librerias necesarias
import os
import json
import re
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
# Cargar variables de entorno
load_dotenv(override=True)
```

```python
# Verificar que ha cargado la api key
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"La clave API de OpenAI existe y empieza por {openai_api_key[:8]}")
else:
    print("La clave API de OpenAI no existe.")
```

```python
# Definir el prompt o cuestión
question = """
España detecta un brote de Peste Porcina Africana en una explotación cercana a zonas con jabalíes.
No existe vacuna eficaz y el impacto económico puede ser elevado.

Propón un plan de actuación que incluya:
1) Medidas sanitarias inmediatas
2) Medidas económicas
3) Gestión de fauna silvestre
4) Coordinación institucional
5) Comunicación pública

Sé claro y estructurado. Añade alguna medida que consideres innovadora aprovechando el uso del IA que marque diferencia de los planes ya establecidos hasta la fecha.
"""
```

```python
# Crear listas donde se almacena las  respuestas de los diferentes modelos además de la definición de mensajes para los modelos
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
```

```python
# Empleo de modelo de open AI gpt-4-mini (Modelo frontier de pago más barato)
openai = OpenAI()
model_name="gpt-4o-mini"
response = openai.chat.completions.create(
    model=model_name,
    messages=messages,
)
answer = response.choices[0].message.content
# Disposición del plan en markdwon
display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Empleo de modelos en abierto descargados a través de ollama y utilizando el end point de OpenAI
# Estos modelos se ejecutan en nuestra máquina por lo que tenemos que ser conscientes de los recursos limitados.
# LLAMA 3.2

ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# GEMMA 3:4b
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "gemma3:4b"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# QWEN 3:4b
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "qwen3:4b"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Guardamos la diferentes respuestas en un archivo zip
for competitor, answer in zip(competitors, answers):
    print(f"Competidor: {competitor}\n\n{answer}")
```

```python
# Juntamos todas las respuestas y las anonizamos
together = ""
for index, answer in enumerate(answers):
    together += f"#Respuesta del competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
# Prompt del juez
judge = f"""Estás juzgando una competición entre {len(competitors)} competidores.
A cada modelo se le ha pedido diseñar un plan para gestionar un brote de peste porcina africana (PPA) en España con la siguiente pregunta:

{question}

Tu trabajo es evaluar cada respuesta según estos criterios globales:
- Eficacia sanitaria para contener y erradicar la PPA.
- Viabilidad técnica y económica de las medidas propuestas.
- Impacto económico, social y ambiental.
- Claridad y coherencia del plan.

Clasifica las respuestas en orden de mejor a peor, teniendo en cuenta todos los criterios de forma conjunta.

Responde con JSON, y solo JSON, con el siguiente formato obligatoriamente:

{{"resultados": ["número del mejor competidor", "número del segundo mejor", "número del tercer mejor", ...]}}

Aquí están las respuestas de cada competidor:

{together}

Ahora responde con el JSON con el orden clasificado de los competidores, nada más. No incluyas formato markdown ni bloques de código."""
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# JUEZ OPENA AI O3
openai = OpenAI()
response = openai.chat.completions.create(
    model="o3-mini",
    messages=judge_messages,
)
results_o3 = response.choices[0].message.content
print(results_o3)

# VEREDICTO o3
print("\nVEREDICTO gpt-o3-mini")
results_dict_o3 = json.loads(results_o3)
ranks = results_dict_o3["resultados"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```

```python
# JUEZ QWEN 3.4b
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "qwen3:4b"

response = ollama.chat.completions.create(model=model_name, messages=judge_messages)
results_q3 = response.choices[0].message.content
print(results_q3)
```

```python
print("\nVEREDICTO MODELO QWEN3:4B")

# results_model es el string que devuelve el modelo, por ejemplo: '{"order": [4, 3, 2, 1]}'
results_dict = json.loads(results_q3)

order = results_dict["order"]   # [4, 3, 2, 1]

for index, comp_num in enumerate(order):
    # comp_num es 4,3,2,1 → los pasamos a índice 0-based
    competitor = competitors[comp_num - 1]
    print(f"Rank {index + 1}: {competitor}")
```
